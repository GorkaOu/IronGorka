```python
import pandas as pd  
import numpy as np
```

```python
import pymysql                        
from sqlalchemy import create_engine  
from getpass import getpass
password = getpass()
connection_string = 'mysql+pymysql://root:'+password+"@localhost/mid_project"
engine = create_engine(connection_string)
```

```python
Airbnb = pd.read_csv('Airbnb.csv')
Amenities = pd.read_csv('Amenities.csv')
Bicycle = pd.read_csv('Bicycle.csv')
Rentpricevo = pd.read_csv('Rentpricevo.csv')
cultural_events = pd.read_csv('cultural-events.csv')
population = pd.read_csv('population.csv')
restaurants = pd.read_csv('restaurants.csv')
transport = pd.read_csv('transport.csv')
Buypricevo = pd.read_csv('Buypricevo.csv')
```

```python
AirbnbSQL = "airbnb"
Airbnb.to_sql(AirbnbSQL, con=engine, if_exists='replace', index=False)
```

```python
AmenitiesSQL = "amenities"
Amenities.to_sql(AmenitiesSQL, con=engine, if_exists='replace', index=False)
```

```python
BicycleSQL = "bicycle"
Bicycle.to_sql(BicycleSQL, con=engine, if_exists='replace', index=False)
```

```python
RentpricevoSQL = "rentpricevo"
Rentpricevo.to_sql(RentpricevoSQL, con=engine, if_exists='replace', index=False)
```

```python
cultural_eventsSQL = "cultural_events"
cultural_events.to_sql(cultural_eventsSQL, con=engine, if_exists='replace', index=False)
```

```python
populationSQL = "population"
population.to_sql(populationSQL, con=engine, if_exists='replace', index=False)
```

```python
restaurantsSQL = "restaurants"
restaurants.to_sql(restaurantsSQL, con=engine, if_exists='replace', index=False)
```

```python
transportSQL = "transport"
transport.to_sql(transportSQL, con=engine, if_exists='replace', index=False)
```
