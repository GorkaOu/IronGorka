```python
Aec['TIPO'].isna().sum()
```

```python
# Translate column names to English.
Aec= Aec.rename(columns={'ID-EVENTO':'Event-ID',
                         'TITULO':'Title', 
                         'DISTRITO-INSTALACION' :'District',
                         'COORDENADA-X' :'X-coordinate',
                         'COORDENADA-Y' :'Y-coordinate',
                         'LATITUD' :'Latitude',
                         'LONGITUD' :'Length',
                         'TIPO' :'Type'})
```

```python
# Aec['X-coordinate'].astype(int)
Aec['District']= Aec['District'].str.lower()
Aec
```

```python
Aec.to_csv('cultural-events.csv', index = False)
```

# Airbnb

```python
Airbnb = pd.read_excel("Airbnb Kaggle.xlsx")
Airbnb.head()
```

```python
# Here I need the information about the numer of Airbnb for each District and the price just to know the area where most tourists are accommodated
Airbnb = Airbnb.drop(['host_id','host_name','name', 'neighbourhood', 'minimum_nights', 'number_of_reviews','last_review','reviews_per_month','calculated_host_listings_count', 'availability_365', 'Unnamed: 16'], axis=1)
```

```python
# In this case there are few NaN values in a big Dataset, so I will drop them.
Airbnb.isna().sum()
```

```python
# Drop all the NaN values, we have a big dataset.
Airbnb = Airbnb.dropna()
```

```python
Airbnb= Airbnb.rename(columns={'id':'airbnb-ID', 'neighbourhood_group' : 'district'})
```

```python
Airbnb ['district'].value_counts()
```

```python
street_list = [] 
for street in Airbnb ['district'].unique():
    if len(Airbnb[Airbnb['district'] == street]) < 10:
        street_list.append(street)
```

```python
Airbnb = Airbnb[~Airbnb['district'].isin(street_list)]
```

```python
def remove_accent(x):
    x = x.replace("Ã","i")
    x = x.replace("i¡","a")
    return x
```

```python
Airbnb['district'] = Airbnb['district'].apply(remove_accent)
Airbnb['district'].value_counts()
```

```python
Airbnb['district']= Airbnb['district'].str.lower()
Airbnb
```

```python
Airbnb.to_csv('Airbnb.csv', index = False)
```

# Bicycle Stops

```python
Bicycle = pd.read_excel("Aparca bicis.xlsx")
Bicycle
```

```python
Bicycle = Bicycle.drop(['DESCRIPTIO,C,105','MINTBARRIO,C,10', 'NOMBRE_BAR,C,50', 'INSTALLDAT,D','MINTTIPOVI,C,15','MINTNOMBRE,C,254','MINTNUMERO,C,25','MINTNDP,C,20','MODELO,C,254'], axis=1)
```

```python
Bicycle = Bicycle.rename(columns={'MXASSETNUM,C,12':'Bicyclestop-ID',
                         'MINTDISTRI,C,20':'District-ID', 
                         'NOMBRE_DIS,C,25':'District',
                         'ESTADO,C,50' :'Condition'})
```

```python
Bicycle['Condition'].value_counts()
```

```python
def cleanOperation(x):
    if 'OPERATIVO' in x:
        return "OPERATIVE"
    elif 'ACTIVO' in x:
        return "OPERATIVE"
    else:
        return x

Bicycle['Condition'] = list(map(cleanOperation, Bicycle['Condition']))
```

```python
#I Will drop them now because 100% are operative.
Bicycle = Bicycle.drop(['Condition'], axis=1)
Bicycle['District']= Bicycle['District'].str.lower()
Bicycle
```

```python
Bicycle.to_csv('Bicycle.csv', index = False)
```

# Amenities

```python
markets = pd.read_excel("mercados.xlsx")
sportscenter = pd.read_excel("Polideportivos.xlsx")
University = pd.read_excel("universidades-educacion.xlsx")
display (markets)
display (sportscenter)
display (University)
```

```python
markets = markets.drop (['PK'],axis = 1 )
markets = markets.rename(columns={'NOMBRE':'Name', 'COD-DISTRITO' : 'District-ID', 'DISTRITO': 'District', 'COORDENADA-X' : 'Cordenate-X', 'COORDENADA-Y':'Cordenate-Y', 'LATITUD': 'latitude', 'LONGITUD':'longitude' })
```

```python
markets['District-ID']= markets['District-ID'].fillna(17)
markets['Cordenate-X'] = markets['Cordenate-X']. fillna(0)
markets['Cordenate-Y'] = markets['Cordenate-Y']. fillna(0)
markets['District']= markets['District'].str.lower()
```

```python
sportscenter = sportscenter.drop (['PK', 'Unnamed: 9'],axis = 1 )
sportscenter = sportscenter.rename(columns={'NOMBRE':'Name', 'COD-DISTRITO' : 'District-ID', 'DISTRITO': 'District', 'COORDENADA-X' : 'Cordenate-X', 'COORDENADA-Y':'Cordenate-Y', 'LATITUD': 'latitude', 'LONGITUD':'longitude','TIPO': 'Type'})
sportscenter['District']= sportscenter['District'].str.lower()
```

```python
# I want to drop from this column '/contenido/entidadesYorganismos/CentrosEspacio..'
Type = []
for c in sportscenter ['Type']:
    Type.append(c.split ('/')[4: ])
sportscenter['Type'] = Type
```

```python
University = University.drop (['PK', 'Unnamed: 9'],axis = 1 )
University = University.rename(columns={'NOMBRE':'Name', 'COD-DISTRITO' : 'District-ID', 'DISTRITO': 'District', 'COORDENADA-X' : 'Cordenate-X', 'COORDENADA-Y':'Cordenate-Y', 'LATITUD': 'latitude', 'LONGITUD':'longitude', 'TIPO': 'Type' })
```

```python
University['District']= University['District'].str.lower()
```

```python
Amenities = pd.concat([markets, sportscenter, University], ignore_index=True, axis=0)
```

```python
Amenities
```

```python
Amenities.to_csv('Amenities.csv', index = False)
```
